# **Fundamentação Teórica**

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;O Processamento Digital de Imagens (PDI) não é uma tarefa simples, na realidade envolve um conjunto de tarefas interconectadas (Fig. 1). Tudo se inicia com a captura de uma imagem, a qual, normalmente, corresponde à iluminação que é refletida na superfície dos objetos, realizada através e um sistema de aquisição. Após a captura por um processo de digitalização, uma imagem precisa ser representada de forma apropriada para tratamento computacional. Imagens podem ser representadas em duas ou mais dimensões.
</p>(QUEIROZ; GOMES, 2001).<br><br>

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;Para elaboração deste trabalho iremos utilizar apenas figuras bidimensionais(2D), Cada ponto na grade bidimensional que representa a imagem digital é denominado elemento de imagem ou pixel. Na Fig. 2, apresenta-se a notação matricial usual para a localização de um pixel no arranjo de pixels de uma imagem bidimensional. O primeiro índice denota a posição da linha, m, na qual o pixel se encontra, enquanto o segundo, n, denota a posição da coluna. Se a imagem digital contiver M linhas e N colunas, o índice m variará de 0 a M-1, enquanto n variará de 0 a N-1. Observe-se o sentido de leitura (varredura) e a convenção usualmente adotada na representação espacial de uma imagem digital.</p>(QUEIROZ; GOMES, 2001).<br><br>

<p align = "center"><img src="https://i.ibb.co/fxzJWB8/image4.png"></p>

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;Os pixels de uma imagem podem ser compostos de apenas um único valor, o qual irá retorna uma imagem com tons de cinza, onde tal valor determinará apenas a intensidade de luminância naquele pixel, mas também pode ser um conjunto de 3 valores(RGB) que em conjunto determinam as cores e a luminância da imagem.</p>

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;Podemos entender RGB como a abreviatura de um sistema de cores aditivas em que o Vermelho (Red), o Verde (Green) e o Azul (Blue) são combinados de várias formas de modo a reproduzir um largo espectro cromático. O propósito principal do sistema RGB é a reprodução de cores em dispositivos eletrônicos como monitores de TV e computador, retroprojetores, scanners e câmeras digitais, assim como na fotografia tradicional.</p>
<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;As cores são atingidas com a mistura das três cores primárias em um número definido numa escala de 0 a 255. Na mistura das três o valor máximo (255) atinge a cor branca e o valor mínimo (0) resulta na cor preta.</p>

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;Já YIQ podemos entender como o espaço de cores usado pelo sistema de TV em cores NTSC, empregado principalmente na América do Norte e Central e no Japão. Eu defendo in-phase, enquanto Q significa quadratura, referindo-se aos componentes usados ​​na modulação de amplitude em quadratura. Algumas formas de NTSC agora usam o espaço de cor YUV, que também é usado por outros sistemas, como o PAL.</p>

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;Ao converter uma imagem do sistema de cores RGB para o YIQ é necessário aplicar três fórmulas que extraem a luminância das cores denotada por Y e as crominâncias da imagem que são armazenadas em I e Q.</p><br>

**<p align = "center">Y = 0.299R + 0.587G + 0.114B<br>
I = 0.596R - 0.274G - 0.322B<br>
Q = 0.211R - 0.523G + 0.312B</p><br>**

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;Já para a conversão contrária de YIQ e RGB é necessário aplicar outras três fórmulas para fazer a transcrição de volta da informação não-visual que o sistema YIQ representa para o RGB.</p><br>

**<p align = "center"> R = 1.000Y + 0.956I + 0.621Q<br>
G = 1.000Y - 0.272I - 0.647Q<br>
B = 1.000Y - 1.106I + 1.703Q</p><br>**

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;Em PDI existem uma grande diversidade de filtros para processamneto de imagens, abaixo iremos conhecer melhor alguns filtros os quais foram utilizados durante a elaboração deste trabalho.</p><br>

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;As técnicas de filtragem são transformações da imagem pixel a pixel, que não dependem apenas do nível de cinza de um determinado pixel, mas também do valor dos níveis de cinza dos pixels vizinhos.
O processo de filtragem é feito utilizando matrizes denominadas máscaras, as quais são aplicadas sobre a imagem. <p><br>

<p align = "center"><img src="https://i.ibb.co/DgYRWSz/img00001.gif"></p>
<p align = "center">Fig. 3 - Máscara de 3 linhas por 3 colunas com centro na posição (2,2).</p><br>

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;A aplicação da máscara com centro na posição (i,j), sendo i o número de uma dada linha e j o número de uma dada coluna sobre a imagem, consiste na substituição do valor do pixel na posição (i,j) por um novo valor que depende dos valores dos pixels vizinhos e dos pesos da máscara, gerando uma nova imagem com a eliminação das linhas e colunas iniciais e finais da imagem original.</p><br>

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;Os filtros espaciais podem ser classificados em passa-baixa, passa-alta ou passa-banda. Os dois primeiros são os mais utilizados em processamento de imagens. O filtro passa-banda é mais utilizado em processamentos específicos, principalmente para remover ruídos periódicos.</p><br>

<font size="4">**Histograma:**</font> <br />

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp; O histograma de uma imagem, descreve a distribuição estatística dos níveis de cinza em pixels com cada nível.<br><br>

<font size="4">**Pivô:**</font> <br />

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;É a célula ou posição onde será realizada a fixação da máscara para aplicação dos filtros ou processos.<br><br>

<font size="4">**Correlação:**</font> <br />

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;Em estatística, a correlação entre dois sinais, ou variáveis, é uma medida padronizada de relação(similaridade) entre os sinais.<br><br>

<font size="4">**Filtro Negativo de RGB:**</font> <br />

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;Este filtro faz com que a cor de cada pixel da imagem original se transforme na cor inversa (ex: pixel branco se transforma em pixel preto). A cor inversa é o valor da subtração entre 255 e o valor RGB. Ex: para um amarelo com valor RGB 255, 240, 0, sua cor inversa é o RGB 0, 15, 255.</p><br>

<font size="4">**Filtro Negativo de YIQ:**</font> <br />

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;Do mesmo modo existe a possibilidade de aplicar o negativo de modo a não alterar a cor do pixel, O negativo deve se aplicado em Y que representa a luminância, o intervalo de valores do canal Y é o mesmo que os canais RGB(0-250), ou seja, após aplicamos a conversão de um pixel RGB para YIQ devemos efetuar y = 255 - y, mantendo os canais I e Q sem alterações, após deve-se converter de volta para RGB para obtemos o negativo de um com sem alteração da sua cor.</p><br>

<font size="4">**Filtro da Média:**</font> <br />

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;O filtro da média percorre toda a imagem, utilizando de uma matriz pre-definida, realizando a troca do quadro na posição central da matriz, na imagem, com o resultado do cálculo da média dos elementos. Quanto maior o filtro, maior a influencia dos vizinhos no cálculo da posição central.</p><br>

<font size="4">**Filtro da Médiana:**</font> <br />

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;O filtro de mediana após enfileirar os termos, através de uma matriz pre-definida, seleciona o termo central ou cálculo dos centros(em casos da quantidade de termos serem pares) e coloca na posição do pixel central. Eles são indicados para reduzir bluring
(desfoque).</p><br>

<font size="4">**Filtro de Sobel:**</font> <br />

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;O filtro Sobel é um operador usado no processamento de imagens para detecção de bordas . Este é um dos operadores mais simples.<br>

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;O operador calcula o gradiente da intensidade de cada pixel. Isso indica a direção da mudança mais forte de claro para escuro, bem como a taxa de mudança nessa direção. Conhecemos então os pontos de mudança repentina na luminosidade, provavelmente correspondendo às arestas, bem como a orientação dessas arestas.<br>

<p align = "justify">&nbsp;&nbsp;&nbsp;&nbsp;Em termos matemáticos, o gradiente de uma função de duas variáveis ​​(aqui a intensidade em função das coordenadas da imagem) é um vetor de dimensão 2 cujas coordenadas são as derivadas de acordo com as direções horizontal e vertical. Em cada ponto, o gradiente aponta na direção da maior mudança de intensidade e seu comprimento representa a taxa de mudança nessa direção. O gradiente em uma zona de intensidade constante é, portanto, zero. Em um contorno, o gradiente cruza o contorno, das intensidades mais escuras às intensidades mais claras.</p><br>

# **Questão 1**

Ao converter uma imagem do sistema de cores RGB para o YIQ é necessário aplicar três fórmulas que extraem a luminância das cores denotada por Y e as crominâncias da imagem que são armazenadas em I e Q. Foram utilizados os valores demostrados aula 2, para realizar a conversão RGB to YIQ.

Definição da função para conversão de RGB para YIQ:

```python
def rgb_para_yiq(r,g,b):
  
    # Utilizando os valores informados na aula 2,
    # Foi aplicado esse procedimento para realizar a conversão
    y = (((0.299 * r) + (0.587 *  g) + (0.114  * b)))
    i = (((0.596 * r) + (-0.274 * g) + (-0.322 * b)))
    q = (((0.211 * r) + (-0.523 * g) + (0.312  * b)))

    return (y,i,q)
```

A função utilizada, faz a leitura como 
BGR ao invés de RGB, dessa forma, o vetor B deve estar na posição 0:

```python
# Inicializando BGR
b = image[:,:,0] 
g = image[:,:,1]  
r = image[:,:,2]

# Definindo então variavel YIQ como
# O retorno dado pela função rgb_para_yiq
YIQ = rgb_para_yiq(r,g,b)

# Repassado os valores de retorno para
# Cada variável correspondente ao array
q = np.array(YIQ[0])
i = np.array(YIQ[1])
y = np.array(YIQ[2])

# Criando array para receber nova imagem YIQ
conversao_yiq = []
conversao_yiq.append(q)
conversao_yiq.append(i)
conversao_yiq.append(y)
conversao_yiq = np.array(conversao_yiq)

# Invertendo array de BGR para exibição RGB
conversao_yiq = np.moveaxis(conversao_yiq, 0, 2)

# Exibindo Imagem
exibir_img(conversao_yiq)
```

Verificar o limite RGB:

```python
#Definição da função para verificação do limite de RGB
def VerificacaoLimite_RGB(r,g,b):

    # Verifica se há valores acima de 255 e os coloca para 255
    if r > 255: 
        r = 255
    if g > 255:
        g = 255
    if b > 255:
        b = 255

    # Verifica se há valores negativos e os coloca para 0
    if r < 0.0:
        r = 0.0
    if g < 0.0:
        g = 0.0
    if b < 0.0:
        b = 0.0

    return r,g,b
```

Função para realizar conversão de retorno do YIQ para RGB:

```python
def yiq_para_rgb(y,i,q):
  
    # Utilizando os valores informados na aula 2,
    # Foi aplicado esse procedimento para realizar a conversão
    r = (((1 * y) + (0.956 *  i) + (0.621  * q)))
    g = (((1 * y) + (-0.272 * i) + (-0.647 * q)))
    b = (((1 * y) + (-1.106 * i) + (1.703  * q)))

    return (r,g,b)
```

Conversão YIQ para RGB:

```python
def Conversao_YIQ_para_RGB(conversaoyqi, r, g, b):
    #Cria um array para receber imagem convertida
    Imagem_Convertida = []

    # Percorre a imagem YIQ
    for lista_row in conversaoyqi:
        array = []
        
        for yiq in lista_row:
            y = yiq[0]
            i = yiq[1]
            q = yiq[2]

            # Conversão e Verificar Limite
            r,g,b = yiq_para_rgb(y,i,q)
            VerificacaoLimite_RGB(r,g,b)

            # Arredondando os valores de RGB
            r = int(round(r))
            g = int(round(g))
            b = int(round(b))
            yiq2 = (r,g,b)
            array.append(yiq2)
            
        Imagem_Convertida.append(array)
      
    return Imagem_Convertida
```

Convertendo a Imagem

```python
# Converte YIQ para RGB
img_Conversao_Para_RGB = Conversao_YIQ_para_RGB(conversao_yiq, r, g, b)

# Exibindo Imagem
exibir_img(img_Conversao_Para_RGB)
```

Função para criar o negativo:

```python
def negativo(img, channel):

    # Cria cópia e salva altura e Largura
    copia = img.copy()
    altura, largura, canais = copia.shape

    # Percorre a imagem e aplica a fórmula do
    # Filtro negativo de acordo com a flag recebida
    for i in range(0, altura - 1): 
        for j in range(0, largura - 1): 

            pixel = copia[i, j] 
            if channel == 'r' or channel == 'all':
                pixel[0] = 255 - pixel[0] 
            if channel == 'g' or channel == 'all':
                pixel[1] = 255 - pixel[1] 
            if channel == 'b' or channel == 'all':
                pixel[2] = 255 - pixel[2]
            if channel == 'y':
                pixel[0] = 255 - pixel[0]

            copia[i, j] = pixel 
            
    return copia
```

# **Questão 3**

Acessar arquivo do filtro:

```python
def filtro_do_txt(nome_arquivo):

    # Inicializando variáveis
    m = []
    arquivo = open(nome_arquivo, "r")

    # Armazena as linhas
    linhas = [l.split(" ") for l in arquivo.readlines()]

    for l in linhas:
        m.append([float(el.rstrip()) for el in l if len(el) > 1])

    return m
```

Expansão por histograma:

```python
def expansao_por_histograma(img, L):

    # Cria cópia e salva altura e Largura
    imagem_antiga = np.copy(img)
    imagem_nova = np.ndarray(imagem_antiga.shape)

    # Armazena o Rmax e Rmin
    r_min = np.min(imagem_antiga)
    r_max = np.max(imagem_antiga)

    # Define o fator da expansão
    fator = (L-1)/(r_max-r_min)

    # Cria uma nova imagem depois de realizar a operação com o fator para normalização
    imagem_nova[:,:,:] = (imagem_antiga[:,:,:] - r_min) * fator
    
    return imagem_nova.astype(int)
```

Aplicar Extensão:

```python
def aplica_extensao_0(img, m, n):

    # Inicializando Variáveis
    pivo = int((max(m, n))/2)
    linha0 = [[0,0,0]]*(len(img[0])+(2*pivo))
    linhas = []

    # Percorre a imagem e atribui o valor as linhas
    for l in img:
        l_0 = [[0,0,0]]*(pivo)
        linhas.append(l_0+l+l_0)

    # Realiza as operações na imagem com o pivo
    img = ([linha0]*pivo) + linhas + ([linha0]*pivo)

    return img
```

Criar Correlacão:

```python
def correlacao(h, m, n, img, pixel_atual, offset):

    # Inicializando variáveis
    i, j = pixel_atual
    novo_pixel = [0, 0, 0]

    pivo = open("www.dropbox.com/s/fygaal8g55yderd/pivo5.txt?dl=0")
    pivovalor = pivo.readlines()
    linha_pivo = 0
    pivm = int(pivovalor[linha_pivo].split()[0])
    pivn = int(pivovalor[linha_pivo].split()[1])

    if(pivm > m and pivn >n):
        # Pivo da posição I    
        pivoI = int((m-1)/2)
        pivo_descerI = pivoI + 1 - m%2

        # Pivo da posição J
        pivoJ = int((n-1)/2)
        pivo_descerJ = pivoJ + 1 - n%2
    else:
        pivoI = pivm
        pivo_descerI = pivoI + 1 - m%2

        pivoJ = pivn
        pivo_descerJ = pivoJ + 1 - n%2
    
    # Incrementa o I e J com o valor do Pivo
    i += pivoI
    j += pivoJ

    # Realiza as operações com os pivôs mais o offset
    novo_pixel[0] = round(np.multiply(h, img[i-pivoI:i+pivo_descerI+1, j-pivoJ:j+pivo_descerJ+1, 0]).sum()) + offset
    novo_pixel[1] = round(np.multiply(h, img[i-pivoI:i+pivo_descerI+1, j-pivoJ:j+pivo_descerJ+1, 1]).sum()) + offset
    novo_pixel[2] = round(np.multiply(h, img[i-pivoI:i+pivo_descerI+1, j-pivoJ:j+pivo_descerJ+1, 2]).sum()) + offset

    return novo_pixel
```

Aplicando o que foi feito

```python
def aplica_correlacao(mascara, img):

    # Cria cópia e salva a mascara e imagem
    mascara_m, mascara_n = np.copy(mascara).shape
    m_ext = np.copy(aplica_extensao_0(img, mascara_m, mascara_n))
    imagem_nova = np.copy(img)
    altura, largura, canais = imagem_nova.shape
    valor_lido = open("www.dropbox.com/s/3v61xyi6yglp6nf/offset.txt?dl=0")
    off = int(valor_lido.read())

    # Aplica a correlação em cada Pixel em altura e comprimento
    for i in range(0, altura):
        for j in range(0, largura):
            imagem_nova[i, j] = correlacao(mascara, mascara_m, mascara_n, m_ext, (i, j), off)

    return np.abs(imagem_nova).tolist()
```

Mediana M x N ímpares 31x35:

```python
def mediana(n, m, imagem):

    # Armazena altura, largura e n° de canais
    # Cria matriz do tamanho da imagem original para guardar a saida
    altura, largura, canais = imagem.shape
    filtroMediana = np.zeros_like(imagem)

    # Cria matriz extendida e preenche
    # Com os valores da imagem original
    colunas = int((m) / 2)
    linhas = int((n) / 2)
    matrizExt = np.zeros((altura+2*colunas, largura+2*linhas, canais)) #Alterando tamanho do filtro para tamanho da imagem
    matrizExt[colunas:altura+colunas, linhas:largura+linhas] = imagem #Percorrendo imagem que receberá o filtro

    # Percorre os pixels e calcula a mediana
    for y in range (colunas , altura+colunas):
        for x in range (linhas , largura +linhas):
            
            #Pega o pixel central da coluna e linha
            valoresY = matrizExt[y-colunas : y+colunas + 1, x-linhas : x+linhas + 1, 0]
            valoresI = matrizExt[y-colunas : y+colunas + 1, x-linhas : x+linhas + 1, 1]
            ValoresQ = matrizExt[y-colunas : y+colunas + 1, x-linhas : x+linhas + 1, 2]
          
            medianaY = int(np.median(valoresY))
            medianaI = int(np.median(valoresI))
            medianaQ = int(np.median(ValoresQ))
             
            # Salva pixel com mediana aplicada na matriz criada 
            filtroMediana[y -colunas,x -linhas,0] = medianaY
            filtroMediana[y -colunas,x -linhas,1] = medianaI
            filtroMediana[y -colunas,x -linhas,2] = medianaQ

    return filtroMediana
    
imagem_Mediana = mediana(9,13,conversao_yiq)
```

Fazendo o matching das imagens:

```python
# Inicializando as variáveis
wonan_eye = cv2.imread('wonan_eye.png', 0)
wonan = cv2.imread('wonan.png', 0)

#Armazena as dimensões da imagem
w, h = wonan_eye.shape[::-1]

# Obtem a região mais correlacionada retornando uma imagem com tons de cinza
res = cv2.matchTemplate(wonan, wonan_eye, eval('cv2.TM_CCORR_NORMED'))

# Obtem os valores globais mínimos e máximos da correlação
valor_min, valor_max, left_min, left_max = cv2.minMaxLoc(res)

# Define o box
top_left = left_max
box = (top_left[0] + w, top_left[1] + h)

# Exibindo resultado do Matching da área de maior correlação
plt.subplot(121),plt.imshow(res,cmap = 'gray')
plt.title('Primeiro Matching'), plt.xticks([]), plt.yticks([])

#Cria uma variável vazia
#Armazena nela uma nova imagem com a área mais correlacionada marcada
wonan2 = ''
wonan2 = cv2.rectangle(wonan, top_left, box, 0, -1)

#exibi a imagem do passo anterior
plt.subplot(122),plt.imshow(wonan2,cmap = 'gray')
plt.title('Primeiro Ponto detectado'), plt.xticks([]), plt.yticks([])
plt.show()

#cria uma variavel vazia
#Armazena nela a segunda região mais correlacionada
res2 = ''
res2 = cv2.matchTemplate(wonan2, wonan_eye, eval('cv2.TM_CCORR_NORMED'))

# Obtem os valores globais mínimos e máximos da correlação
valor_min, valor_max, left_min, left_max = cv2.minMaxLoc(res2)

# Define o box
top_left = left_max
box2 = (top_left[0] + w, top_left[1] + h)

# Exibindo resultado do Matching da segunda região mais correlacionada
plt.subplot(121),plt.imshow(res2,cmap = 'gray')
plt.title('Segundo Matching'), plt.xticks([]), plt.yticks([])

#Cria uma variável vazia
#Armazena nela uma nova imagem com a segunda área mais correlacionada marcada
wonan3 = ''
wonan3 = cv2.rectangle(wonan2, top_left, box2, 0, -1)

# #exibi a imagem do passo anterior
plt.subplot(122),plt.imshow(wonan3,cmap = 'gray')
plt.title('Segundo ponto detectado'), plt.xticks([]), plt.yticks([])
plt.show()

#cria uma variavel vazia
#Armazena nela a Terceira região mais correlacionada
res3 = ''
res3 = cv2.matchTemplate(wonan3, wonan_eye, eval('cv2.TM_CCORR_NORMED'))

# Obtem os valores globais mínimos e máximos da correlação
valor_min, valor_max, left_min, left_max = cv2.minMaxLoc(res3)

# Define o box
top_left = left_max
box3 = (top_left[0] + w, top_left[1] + h)

# Exibindo resultado do Matching
plt.subplot(121),plt.imshow(res3,cmap = 'gray')
plt.title('Terceiro Matching'), plt.xticks([]), plt.yticks([])

#Cria uma variável vazia
#Armazena nela uma nova imagem com a terceira área mais correlacionada marcada
wonan4 = ''
wonan4 = cv2.rectangle(wonan3, top_left, box3, 255, 0)

#exibi a imagem do passo anterior
plt.subplot(122),plt.imshow(wonan4,cmap = 'gray')
plt.title('Terceiro Ponto detectado'), plt.xticks([]), plt.yticks([])
plt.show()
```
